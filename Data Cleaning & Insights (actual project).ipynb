### 2, Data Cleaning

```python
df.dtypes
```

```python
df.drop_duplicates(inplace = True)
```

```python
df['Age'] = df['Age'].astype(str)
df['SSN'] = df['SSN'].astype(str)
df['Annual_Income'] = df['Annual_Income'].astype(str)
df['Num_of_Loan'] = df['Num_of_Loan'].astype(str)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype(str)
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].astype(str)
df['Outstanding_Debt'] = df['Outstanding_Debt'].astype(str)
df['Credit_Mix'] = df['Credit_Mix'].astype(str)
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].astype(str)
df['Amount_invested_monthly'] = df['Amount_invested_monthly'].astype(str)
df['Payment_Behaviour'] = df['Payment_Behaviour'].astype(str)


df
```

```python
# I use excel filter to find popuplar special charaters
df['Age'] = df['Age'].str.strip("_-")
df['SSN'] = df['SSN'].str.strip("#F%$D@*&")
df['SSN'] = df['SSN'].str.replace(r'^(8)$','',regex = True)
df['Occupation'] = df['Occupation'].str.strip("_______")
df['Annual_Income'] = df['Annual_Income'].str.strip("_")
df['Num_of_Loan'] = df['Num_of_Loan'].str.strip("_")
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].str.strip("_")
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].str.strip("_")
df['Credit_Mix'] = df['Credit_Mix'].str.strip("_")
df['Outstanding_Debt'] = df['Outstanding_Debt'].str.strip("_")
df['Credit_Mix'] = df['Credit_Mix'].str.strip("_")
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].str.replace(r'^(NM)$','Not Mentioned',regex = True)
df['Amount_invested_monthly'] = df['Amount_invested_monthly'].str.replace(r'^(__10000__)$','',regex = True)
df['Payment_Behaviour'] = df['Payment_Behaviour'].str.replace(r'^(!@9#%8)$','',regex = True)


df
```

### SSN column

#### fill real values into NaN values

```python
#replace blanks into NaN values 
df['SSN'] = df['SSN'].replace('', np.nan)
```

```python
df[(df['Name'].isnull()) | (df['SSN'].isnull())]
```

```python
Customer_IDs = df[df['SSN']== 'NaN']['Customer_ID'].values
Customer_IDs
```

```python
ssn1  = df.dropna(subset = 'SSN').set_index('Customer_ID')['SSN'].to_dict()

df['SSN'] = df.apply(lambda row: ssn1.get(row['Customer_ID'],row['SSN']), axis = 1)
```

```python
df.head(10)
```

#### convert data from xxx-yy-zzzz into xxxyyzzzz type and SSN column into int64 type

```python
df['SSN'].isnull().sum()
```

```python
df.SSN = df.SSN.apply(lambda x: int(''.join(x.split('-'))) if pd.notna(x) else x).astype(np.int64)


#Or we can use this replacement when SSN column has no NaN values

#df['SSN'] = df['SSN'].apply(lambda x: x.replace('-',''))
#df['SSN'] = df['SSN'].astype('int64')
#df
```
