#### This column has some values of the form '0 years and x months', so i would be better to convert it into only months

```python
def month_converter(x):
    if pd.notnull(x):
        year = int(x.split(' ')[0])
        month = int(x.split(' ')[3])
        return (year*12) + month
    else:
        return x
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].apply(lambda x: month_converter(x)).astype(np.float64)
```

```python
customer_id1 = df[(df['Credit_History_Age'].isna())]['Customer_ID'].values

# groupby Customer_ID and interpolate Credit_History_Age values 
df['Credit_History_Age'] = df.groupby('Customer_ID')['Credit_History_Age'].transform(
    lambda x: x.interpolate(method='index', limit_direction='both') if x.count() > 1 else x)
```

### Payment_behaviour

```python
df[df['Payment_Behaviour'].isnull()]
```

```python
# replace blanks into NaN values
df['Payment_Behaviour'] = df['Payment_Behaviour'].replace('',np.nan)
```

```python
def customer_payment(series):
    mode_values = series.dropna().mode()
    if mode_values.empty:
        return np.nan
    return mode_values.iloc[0]
```

```python
payment_mode = df.groupby('Customer_ID')['Payment_Behaviour'].transform(customer_payment) 

df['Payment_Behaviour'] = df['Payment_Behaviour'].fillna(payment_mode)
```
